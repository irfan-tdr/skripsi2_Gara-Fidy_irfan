```python
import numpy as np
import pandas as pd
import time
import datetime
from datetime import date
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import plotly
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
from IPython.display import HTML
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy import stats
from statsmodels.stats import weightstats as stests
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, classification_report, make_scorer, precision_score
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, accuracy_score, confusion_matrix, recall_score
```

```python
columns = ['app_name', 'category', 'rating', 'reviews', 'installs', 'sizes', 'prices', 'content_rating', 'last_updated', 'minimum_version', 'latest_version','month_last_update','years_last_update']
load_data = pd.read_csv('datasetGpsSkripsi02.csv')
load_data.columns = columns
load_data.head()
```

```python
# Removing NaN values
load_data = load_data[pd.notnull(load_data['latest_version'])]
load_data = load_data[pd.notnull(load_data['reviews'])]
load_data.isnull().sum()
```

```python
# to be used: category, rating, installs,sizes, prices, content_rating

load_data = load_data[load_data.sizes.str.contains('\d')]
load_data.sizes[load_data.sizes.str.contains('k')] = "0."+load_data.sizes[load_data.sizes.str.contains('k')].str.replace('.','')
load_data.sizes = load_data.sizes.str.replace('k','')
load_data.sizes = load_data.sizes.str.replace('M','')
load_data.sizes = load_data.sizes.str.replace(',','')
load_data.sizes = load_data.sizes.str.replace('+','')
load_data.sizes = load_data.sizes.astype(float)


load_data = load_data[load_data.installs.str.contains('\+')]
load_data.installs = load_data.installs.str.replace('+','')
load_data.installs = load_data.installs.str.replace(',','')
load_data.installs.astype(int)

load_data.prices  = load_data.prices.str.contains('1|2|3|4|5|7|8|9').replace(False, 0)

load_data = load_data[load_data.applymap(np.isreal).reviews]
load_data.reviews = load_data.reviews.astype(float)

load_data = load_data[load_data.rating.str.contains('\d') == True]
load_data.rating = load_data.rating.astype(float)

load_data.category.unique()
```

```python
data_category = load_data.category.fillna('Unknown')
games = load_data[data_category.str.contains('GAME', regex=False)]
other = load_data[~data_category.str.contains('GAME', regex=False)]

z_Rating = np.abs(stats.zscore(games.rating))
games = games[z_Rating < 3]
z_Reviews = np.abs(stats.zscore(games.reviews))
games = games[z_Reviews < 3]

z_Rating2 = np.abs(stats.zscore(other.rating))
other = other[z_Rating2 < 3]
z_Reviews2 = np.abs(stats.zscore(other.reviews))
other = other[z_Reviews2 < 3]
```

```python
games_mean = np.mean(games.rating)
games_std = np.std(games.rating)

other_mean = np.mean(other.rating)
other_std = np.std(games.rating)

print('Games mean and std: ', games_mean, games_std)
print('Other categories mean and std: ', other_mean, other_std)

ztest, pval = stests.ztest(games.rating, other.rating, usevar='pooled', value=0, alternative='smaller')
print('p-value: ', pval)
```

```python
data = load_data.drop(['app_name', 'reviews', 'minimum_version', 'latest_version','last_updated'], axis=1)
data.head()
```

```python
category_unique = list(set(data.category))
rating_unique = list(set(data.rating))
install_unique = list(set(data.installs))
size_unique = list(set(data.sizes))
prices_unique = list(set(data.prices))
content_rating_unique = list(set(data.content_rating))
month_last_update_unique = list(set(data.month_last_update))
years_last_update_unique = list(set(data.years_last_update))
data.head()
```

```python
cleared_data = []
all_data = []
for per_row in data.values:
    app = []
    if per_row[0] in category_unique:
        app.append(per_row[0])
    if per_row[1] in rating_unique:
        if float(per_row[1]) > 0 and float(per_row[1]) <= 4:
            app.append('0')
        if float(per_row[1]) > 4:
            app.append('1')
    if per_row[2] in install_unique:
        app.append(per_row[2])
    if per_row[3] in size_unique:
        if float(per_row[3]) > 0 and float(per_row[3]) <= 25:
            app.append('Size <= 25MB')
        if float(per_row[3]) > 25:
            app.append('Size > 25MB')
    if per_row[4] in prices_unique:
        if float(per_row[4]) == 0:
            app.append('GRATIS')
        if float(per_row[4]) == 1:
            app.append('BERBAYAR')
    if per_row[5] in content_rating_unique:
        app.append(per_row[5])
    if per_row[6] in month_last_update_unique:
        app.append(per_row[6])
    if per_row[7] in years_last_update_unique:
        app.append(per_row[7])
    if len(app) == 8:
        cleared_data.append(app)
cleared_data_pd = pd.DataFrame(cleared_data)
data = cleared_data_pd
```

```python
X = data.drop(['rating'], axis=1)
# X['installs'] = X['installs'].astype(str).astype(int)

y=data['rating'].values
y=y.astype('int')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 45)
```

```python
knn=KNeighborsClassifier(n_neighbors=20)
knn.fit(X_train,y_train)
knn_pred=knn.predict(X_test)
knn_cm=confusion_matrix(y_test,knn_pred)
knn_ac=accuracy_score(y_test, knn_pred)

plt.figure(figsize=(10,5))
plt.title("knn_cm")
sns.heatmap(knn_cm,annot=True,fmt="d",cbar=False)
recall_knn = recall_score(y_test, knn_pred,average='weighted')
precision_knn = precision_score(y_test, knn_pred,average='weighted')
RMSE_knn = mean_squared_error(y_test, knn_pred)
print('KNN_Classifier_accuracy:',knn_ac)
print('Recall:',recall_knn)
print('Precision:',precision_knn)
print('RMSE:',RMSE_knn)
print(classification_report(y_test,knn_pred))
```

```python
# search for an optimal value of K for KNN
from sklearn.model_selection import KFold, cross_val_score
# range of k we want to try
k_range = range(1, 21)
# empty list to store scores
k_scores = []

# 1. we will loop through reasonable values of k
for k in k_range:
    # 2. run KNeighborsClassifier with k neighbours
    knn = KNeighborsClassifier(n_neighbors=k)
    # 3. obtain cross_val_score for KNeighborsClassifier with k neighbours
    acc_scores = cross_val_score(knn,X_train, y_train, cv=5, scoring='accuracy')
    # 4. append mean of scores for k neighbors to k_scores list
    k_scores.append(acc_scores.mean())
    acc_cv = acc_scores.mean()
    scores.append(metrics.accuracy_score(y_test, y_pred))
    pre_scores = cross_val_score(knn,X_train, y_train, cv=5, scoring='precision')
    pre_cv = pre_scores.mean()
    rec_scores = cross_val_score(knn,X_train, y_train, cv=5, scoring='recall')
    rec_cv = rec_scores.mean()
    RMSE_scores = -cross_val_score(knn,X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    RMSE_cv = RMSE_scores.mean()
    print('CV accuracy of {}-k: {} | precision {} | recall {} | RMSE {}' . format(k, acc_cv,pre_cv,rec_cv,RMSE_cv))


# print(k_scores)
```

```python
kfold = KFold(n_splits=5, random_state=42)
knn_cv=KNeighborsClassifier(n_neighbors=19)

cv_knn_acc = cross_val_score(estimator=knn_cv,
                                          X=X_train,
                                          y=y_train,
                                          cv=kfold,
                                          scoring='accuracy')
cv_acc = cv_knn_acc.mean()
cv_knn_pre = cross_val_score(estimator=knn_cv,
                                          X=X_train,
                                          y=y_train,
                                          cv=kfold,
                                          scoring='precision')
cv_pre = cv_knn_pre.mean()
cv_knn_rec = cross_val_score(estimator=knn_cv,
                                          X=X_train,
                                          y=y_train,
                                          cv=kfold,
                                          scoring='recall')
cv_rec = cv_knn_rec.mean()

rmse = -cross_val_score(estimator=knn_cv,
                                          X=X_train,
                                          y=y_train,
                                          cv=kfold,
                                          scoring='neg_mean_squared_error')
cv_rmse = rmse.mean()

print('CV Accuracy:',cv_acc)
print('CV Recall:',cv_pre)
print('CV Precision:',cv_rec)
print('CV RMSE:',cv_rmse)
```
